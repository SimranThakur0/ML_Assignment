```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/kaggle/input/image-cnn/240220091259-tesla-cybertruck.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Define horizontal and vertical Sobel kernels
horizontal_kernel = np.array([[-1, -2, -1],
                              [0,  0,  0],
                              [1,  2,  1]])

vertical_kernel = np.array([[-1, 0, 1],
                            [-2, 0, 2],
                            [-1, 0, 1]])
```

```python
plt.imshow(image)
```

```python
horizontal_edges = cv2.filter2D(image, -1, horizontal_kernel)
```

```python
vertical_on_horizontal = cv2.filter2D(horizontal_edges, -1, vertical_kernel)
```

```python
# Display the outputs
plt.figure(figsize=(12, 8))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Horizontal Edges")
plt.imshow(horizontal_edges, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Vertical on Horizontal")
plt.imshow(vertical_on_horizontal, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
